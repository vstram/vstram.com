```python
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
```

```python
i1 = 0
i2 = 10
_lambda = 1
xpoints = np.arange(i1, i2, 0.1)
xrandom =  _lambda*np.exp(-_lambda*np.random.randn(50)) 
exp_dist = _lambda*np.exp(-_lambda*xpoints)

fig, ax = plt.subplots()
ax.plot(xpoints, exp_dist)

ax.axvline(1/_lambda, ls='--', color='r')

n, bins, patches = plt.hist(xrandom, bins=10, density=True, facecolor='g', alpha=0.75)

plt.xlabel('Time Interval[s]')
plt.ylabel('Probability')
plt.title('Exponential Distribution')
plt.axis([i1, i2, 0, max(exp_dist)])
plt.grid(True)
plt.show()
```
